```python
import os
import glob
import shutil
import cv2
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("jsrt_metadata.csv")
```

```python
df = df[["study_id","state"]]
```

```python
df.head()
```

```python
malignant_list = list()
benign_list = list()
```

```python
for index in range(len(df)):
    if df.loc[index,"state"] == "malignant":
        malignant_list.append(df.loc[index,"study_id"])
    elif df.loc[index,"state"] == "benign":
        benign_list.append(df.loc[index,"study_id"])
```

```python
print(len(malignant_list))
print(len(benign_list))
```

```python
df.info()
```

```python
print(malignant_list[0])
print(benign_list[0])
```

```python
DATA_DIR = os.getcwd()
DATA_DIR = DATA_DIR + "\\" + "images"
CATEGORIES=["nodule"]
```

```python
benign_dir = DATA_DIR + "\\" + "benign"
malignant_dir= DATA_DIR + "\\" + "malignant"
```

```python
print(benign_dir)
print(malignant_dir)
```

```python
for category in CATEGORIES:
    path=os.path.join(DATA_DIR,category) 
    for img in os.listdir(path):
        print(img)
        img_path = os.path.join(path,img)
        print(img_path)
        if img in malignant_list:
            shutil.copy(img_path, malignant_dir)
        elif img in benign_list:
            shutil.copy(img_path, benign_dir)
        
```
